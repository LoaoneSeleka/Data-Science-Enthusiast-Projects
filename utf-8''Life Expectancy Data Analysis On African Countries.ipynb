### Feature Engineering

```python
#Get the african dataframe first five records
df_africa.head()
```

```python
#Drop country column as we will not need it when building our model
df_africa = df_africa.drop('Country', axis = 1)
```

```python
#Value counts for status column
df_africa['Status'].value_counts()
```

```python
#One hot encode the status column
df_africa = pd.get_dummies(df_africa, columns=['Status'])
df_africa.head()
```

### Build Model

```python
#Get the correlation values for life expectancy column
df_africa.corr()['Life_expectancy'].sort_values(ascending = False)
```

```python
#Independent variables
X = df_africa[['Income_composition_of_resources','BMI','Schooling','Diphtheria','Polio','Year','GDP','Hepatitis_B'
               ,'Percentage_expenditure','Population','Measles','Total_expenditure','Infant_deaths','Under_five_deaths'
              ,'Thinness _1-19_years','Alcohol','Adult_mortality','HIV/AIDS','Status_Developing']].values
X[0:5]
```

```python
#Imputing missing values with the mean
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean', verbose=0)
imp = imp.fit(X[:,0:19])
X[:,0:19] = imp.transform(X[:,0:19])
```

```python
#Dependent variable
y = df_africa['Life_expectancy'].values
y[0:5]
```

```python
#Divife dataframe into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#Normalizing
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)
```

```python
#Get the coef and intercept
print('Linear Regression coefficient (m)', regressor.coef_)
print('Linear Regression intercept (b)', regressor.intercept_)
```

```python
#Score training set
print(regressor.score(X_train, y_train))
```

```python
#Score test set
print(regressor.score(X_test, y_test))
```

### Test model

```python
#Actual values
y_test[0:20]
```

```python
#Predict values
y_pred = regressor.predict(X_test)
y_pred[0:20]
```

### Evaluate Model

```python
from sklearn.metrics import mean_squared_error, r2_score
#Print variance and RMSE
print('Variance: %s' % r2_score(y_test,y_pred)) #0 is worse, 1 is best
print('RMSE: %s' % mean_squared_error(y_test,y_pred)) # the lower the better
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```

```python
#Scatterplot of truth and predicted values
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred, color = 'red')
plt.ylabel('Model Predictions')
plt.xlabel('True values (Ground Truth)')
plt.title('Multiple Linear Regression Predictions')
plt.show()
```

### K-fold Cross Validation

```python
#Cross validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 5)
accuracies
```

```python
#mean performance of the cross validation
accuracies.mean()
```

```python
#standard deviation of the performance of cross validation
accuracies.std()
```

### Ridge and Lasso Regression

#### Ridge

```python
from sklearn.linear_model import Lasso, Ridge
regressor_ridge = Ridge(alpha = 100)
regressor_ridge.fit(X_train,y_train)

print('Linear Regression coefficient (m)', regressor_ridge.coef_)
print('Linear Regression intercept (b)', regressor_ridge.intercept_)

y_predict = regressor_ridge.predict(X_test)
y_predict[0:5]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
#Print variance and RMSE
print('Variance: %s' % r2_score(y_test,y_predict)) #0 is worse, 1 is best
print('RMSE: %s' % mean_squared_error(y_test,y_predict)) # the lower the better
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
print("Root Mean Squared Error: {}".format(rmse))
```

#### Lasso

```python
from sklearn.linear_model import Lasso, Ridge
regressor_lasso = Ridge(alpha = 50)
regressor_lasso.fit(X_train,y_train)

print('Linear Regression coefficient (m)', regressor_lasso.coef_)
print('Linear Regression intercept (b)', regressor_lasso.intercept_)

predict = regressor_lasso.predict(X_test)
predict[0:5]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
#Print variance and RMSE
print('Variance: %s' % r2_score(y_test,predict)) #0 is worse, 1 is best
print('RMSE: %s' % mean_squared_error(y_test,predict)) # the lower the better
rmse = np.sqrt(mean_squared_error(y_test, predict))
print("Root Mean Squared Error: {}".format(rmse))
```
